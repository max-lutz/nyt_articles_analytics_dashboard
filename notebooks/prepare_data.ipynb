```python
import os
import pandas as pd
from dotenv import load_dotenv
import requests

load_dotenv()
api_key = os.getenv('NYT_API_KEY')
```

```python
parameters = {'api-key':api_key}

response = requests.get("https://api.nytimes.com/svc/archive/v1/2019/1.json", params=parameters)
json = response.json()
```
